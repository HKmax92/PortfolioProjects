## 7. Standardizing NaN values for trees with status of stump and dead to "NOT Applicable".

```python
tree_census_subset['health'].value_counts()
```

```python
mask= ((tree_census_subset['status']=='Stump') | (tree_census_subset['status']=='Dead'))
```

```python
tree_census_subset.loc[mask]=tree_census_subset.loc[mask].fillna('Not Applicable')
```

```python
tree_census_subset[tree_census_subset['status']=='Stump']
```

```python
tree_census_subset.isna().sum()
```

```python
tree_census_subset[tree_census_subset['health'].isna()]
```

### Hint:

So we notice that there is one Alive tree that has 'NaN' value and this value does not actually mean "not a number " but more of 'not important' or 'not observer' at the time ,so we might fill it with the most common value like 'Good' , We do the same for the other null columns.

```python
tree_census_subset['problems'].fillna('None',inplace=True)
tree_census_subset['health'].fillna('Good',inplace=True)
tree_census_subset['spc_latin'].fillna('No Observation',inplace=True)
tree_census_subset['sidewalk'].fillna('NoDamage',inplace=True)
```

```python
tree_census_subset.isna().sum()
```

```python
big_trees =tree_census_subset[(tree_census_subset['stump_diam']>60) | (tree_census_subset['tree_dbh']>60)]
big_trees
```

So we gathered about 245 outliers , so we might cut them out as being a tiny portion of the actual size which is 683788 , but as a fresh exercise we might replace the big diameters with the MIN, Avg or Max diameter .

```python
smaller_trees =tree_census_subset[(tree_census_subset['stump_diam']<=60) & (tree_census_subset['tree_dbh']<=60)]
smaller_trees
```

```python
smaller_trees_alive=smaller_trees[smaller_trees['status']=='Alive']
smaller_trees_dead_or_stump = smaller_trees[(smaller_trees['status']=='Dead') & (smaller_trees['status']=='Stump')]
```

```python
stats_Alive = smaller_trees_alive.groupby('spc_latin')['tree_dbh'].describe().reset_index()[['spc_latin','25%','50%','75%']]
stats_Alive
```

```python
smaller_trees_alive=smaller_trees_alive.merge(stats_Alive, on ='spc_latin', how='left')
smaller_trees_alive
```

```python
mask = smaller_trees_alive['tree_dbh']<smaller_trees_alive['25%']
smaller_trees_alive.loc[mask,'tree_dbh'] = smaller_trees_alive['25%'] 

mask = smaller_trees_alive['tree_dbh']>smaller_trees_alive['75%']
smaller_trees_alive.loc[mask,'tree_dbh'] = smaller_trees_alive['75%']

smaller_trees_alive
```
